```
# prompt: Plot simple Trigonometry function plot in three parts

import matplotlib.pyplot as plt
import numpy as np
# Define the range of angles
x = np.linspace(0, 2*np.pi, 100)

# Choose a trigonometric function to plot
y = np.sin(x)

# Create the plot
plt.subplot(3, 1, 1)
plt.plot(x, y)
plt.xlabel("Angle (radians)")
plt.ylabel("Sine")
plt.title("Sine Function")

# Choose a trigonometric function to plot
y = np.cos(x)

# Create the plot
plt.subplot(3, 1, 2)
plt.plot(x, y)
plt.xlabel("Angle (radians)")
plt.ylabel("Cosine")
plt.title("Cosine Function")

# Choose a trigonometric function to plot
y = np.tan(x)

# Create the plot
plt.subplot(3, 1, 3)
plt.plot(x, y)
plt.xlabel("Angle (radians)")
plt.ylabel("Tangent")
plt.title("Tangent Function")

# Adjust spacing between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

```
# prompt: Create a simple bar graph. Suppose you want to compare the box office of
# the movies released in 2020 in the India

# Import necessary libraries
import matplotlib.pyplot as plt

# Prepare data
movies = ['Movie A', 'Movie B', 'Movie C', 'Movie D', 'Movie E']
box_office = [100, 80, 50, 70, 90]

# Create the bar chart
plt.bar(movies, box_office)

# Add labels and title
plt.xlabel("Movies")
plt.ylabel("Box Office (in crores)")
plt.title("Comparison of Box Office of Movies Released in 2020 in India")

# Show the bar chart
plt.show()
```

```
# prompt: Create simple pie chart. Assume our data represents the result of a survey
# on a favourite programming language

# Import necessary libraries
import matplotlib.pyplot as plt

# Define data
languages = ["Python", "C++", "Ruby", "Java"]
popularity = [504, 337, 415, 280]

# Create the pie chart
plt.pie(popularity, labels=languages, autopct="%1.1f%%")

# Add title
plt.title("Favorite Programming Language")

# Show the pie chart
plt.show()
```

```
# prompt: Let's say you've decided to plot the height of your classmate. Now, create a
# very simple histogram with only 10 values

# Import necessary libraries
import matplotlib.pyplot as plt

# Define the data
height = [160, 170, 180, 165, 175, 185, 168, 172, 182, 178]

# Create the histogram
plt.hist(height, bins=10)

# Add title and labels
plt.title("Histogram of Classmate Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Number of Students")

# Show the histogram
plt.show()
```

```
# prompt: Create a simple 3D graph

import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

# Create the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data
ax.plot(x, y, z)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Simple 3D Graph')

# Show the plot
plt.show()
```
